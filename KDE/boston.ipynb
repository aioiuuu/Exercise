```python
import plotly.graph_objs as go
import numpy as np
from numpy import where,random,array,quantile
from sklearn.preprocessing import scale
from sklearn.neighbors import KernelDensity
```

定义生成随机数据的函数

```python
def prepData(N):
    x=[]
    for i in range(N):
        A = i/1000+random.uniform(-4,3)
        R = random.uniform(-5,10)
        if(R<(-4.9)):
            R = R+(-30)
        x.append([A+R])
    return array(x)
```

绘图

```python
n=500
x = prepData(n)
x_ax = list(range(n))
fig = go.Figure()
fig.add_trace(go.Scatter(x=x_ax,y=x.flatten(),mode='lines',name='line'))
fig.show()
```

使用KernelDensity函数估计概率密度

```python
kern_dens = KernelDensity()
kern_dens.fit(x)
score = kern_dens.score_samples(x)
threshold = quantile(score, .01)
idx = where(score<= threshold)
values = x[idx]
fig = go.Figure()
fig.add_trace(go.Scatter(x=x_ax,y=x.flatten(),mode='lines',name='line'))
fig.add_trace(go.Scatter(x=idx[0],y=values.flatten(),mode='markers',marker_color='red',name='outliers'))
fig.show()
```
